```python
# Import data
import os
import pickle

xrd_data = {}
xrd_pkl = os.listdir('data/plot')
for pkl in xrd_pkl:
    with open('data/plot/' + pkl, 'rb') as f:
        xrd_data[pkl.rstrip('.pkl')] = pickle.load(f)
```

```python
import matplotlib.pyplot as plt
import matplotlib.transforms as mtransforms
from pyxrd import plot_ternary_diagram, rotate_phase_index
```

```python
color_Sn = {
    'amorphous': '#5EB89D',
    'SnSb (R-3m)': '#D1363C',
    'Sb (R-3m)': '#237AA6',
    'SnSb (I4_1/amd)': '#E89C3D',
    'Sn (I4_1/amd)': '#BD448E',
    'GeSn (Fd-3m)': '#43624F',
    'Sb (P6_3/mmc)': '#B695DB',
    'Sn (I4/mmm)': '#63B1E6'
}
```

```python
# Plot GeSbSn
fig, ax = plt.subplot_mosaic([['a)', 'b)'], ['c)', 'd)']], figsize=(10, 7))
plt.tight_layout(pad=2.5)
plt.subplots_adjust(wspace=0.3, left=0.05, right=0.68, top=0.9, bottom=0.1)
for data_name, data in xrd_data.items():
    if data_name == 'GeSbSn_150C':
        plot_ternary_diagram(
            ax=ax['a)'],
            phase_type=data['phase_type'],
            phase_index=data['phase_index'],
            labels=['Sn (%)', 'Ge (%)', 'Sb (%)'],
            color=color_Sn,
            if_show=False,
            if_save=False,
            if_legend=False,
            rotation={
                'GeSn (Fd-3m) + Sn (I4_1/amd)': 300,
                'Sb (R-3m) + SnSb (I4_1/amd)': 0,
                'Sb (R-3m) + SbSe (Pnma)': -90
            },
            title=''
        )
    if data_name == 'GeSbSn_300C':
        plot_ternary_diagram(
            ax=ax['b)'],
            phase_type=data['phase_type'],
            phase_index=data['phase_index'],
            labels=['Sn (%)', 'Ge (%)', 'Sb (%)'],
            color=color_Sn,
            if_show=False,
            if_save=False,
            if_legend=True,
            rotation={
                'GeSn (Fd-3m) + Sn (I4_1/amd)': 300,
                'Sb (R-3m) + SnSb (I4_1/amd)': 0,
                'Sb (R-3m) + SbSe (Pnma)': -90,
                'Sb (P6_3/mmc) + Sn (I4/mmm) + SnSb (R-3m)': 180,
                'SnSb (R-3m) + Sn (I4/mmm) + SnSb (I4_1/amd)': 90,
                'Sb (R-3m) + Sb (P6_3/mmc)': 270,
                'SnSb (R-3m) + Sb (P6_3/mmc)': 270,
                'SnSb (R-3m) + Sn (I4/mmm) + SnSb (I4_1/amd)': 60
            },
            title=''
        )
    if data_name == 'SbSnGe_300C':
        plot_ternary_diagram(
            ax=ax['d)'],
            phase_type=data['phase_type'],
            phase_index=rotate_phase_index(data['phase_index']),
            labels=['Sn (%)', 'Ge (%)', 'Sb (%)'],
            color=color_Sn,
            if_show=False,
            if_save=False,
            if_legend=False,
            rotation={
                'GeSn (Fd-3m) + Sn (I4_1/amd)': 300,
                'Sb (R-3m) + SnSb (I4_1/amd)': 0,
                'Sb (R-3m) + SbSe (Pnma)': -90
            },
            title=''
        )
    if data_name == 'SbSnGe_150C':
        plot_ternary_diagram(
            ax=ax['c)'],
            phase_type=data['phase_type'],
            phase_index=rotate_phase_index(data['phase_index']),
            labels=['Sn (%)', 'Ge (%)', 'Sb (%)'],
            color=color_Sn,
            if_show=False,
            if_save=False,
            if_legend=False,
            rotation={
                'GeSn (Fd-3m) + Sn (I4_1/amd)': 300,
                'Sb (R-3m) + SnSb (I4_1/amd)': 0,
                'Sb (R-3m) + SbSe (Pnma)': -90
            },
            title=''
        )
for label, ax1 in ax.items():
    # label physical distance to the left and up:
    trans = mtransforms.ScaledTranslation(-20/72, 7/72, fig.dpi_scale_trans)
    ax1.text(0.0, 1.0, label, transform=ax1.transAxes + trans,
            fontsize=24, va='bottom', fontfamily='Times New Roman', fontweight='bold')
fig.savefig('figure/GeSbSn_combined.png', dpi=600)
```
